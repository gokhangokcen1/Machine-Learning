```python
# kütüphanelerin eklenmesi
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# m2 başına ev fiyatlarını tahmin etmek istiyoruz. Bu sebeple
# x_train veri seti evimizin boyutu (m2 cinsinden)
# y_train hedef fiyatımız (1000 cinsinden)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m, kaç adet eğitim örneğimiz olduğuna bakıyor, len ile aynı işlev
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
# 2, yazmasının sebebi 2 eğitim örneğimiz olduğu anlamına geliyor. 
# virgülden sonrası da her eğitim örneği için kaç feature olduğunu gösteriyor. 
# bizim her veri setinde yalnızca bir feature'ımız var.
```

```python
# noktaları belirle:
# giriş veri seti, çıkış veri seti, nasıl gösterileceği, rengi
plt.scatter(x_train, y_train, marker='x', c='r')
# plot'un ismi
plt.title("Housing Prices")
# yatak eksenin etiketi
plt.ylabel('Price (in 1000s of dollars)')
# dikey eksenin etiketi
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0] # kaç örneğimiz var 
    f_wb = np.zeros(m) # sıfırlardan oluşan m boyutlu (2x1) matris
    for i in range(m): 
        f_wb[i] = w * x[i] + b # fiyatı yaz
        
    return f_wb
```

```python
w, b = 200, 100
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
def compute_cost(x,y,w,b):
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum += cost
    total_cost = cost_sum / (2 * m)
    return total_cost

print("Cost: {}".format(compute_cost(x_train, y_train, w, b)))

# cost fonksiyonu ne kadar az olursa ağırlıklarımız ve bias değerlerimiz o kadar idealdir.
# 2 örnekli bir veri setinde çalıştığımız için w = 200, b = 150 değerleri bizim için
# 0 cost değerini üretiyor. 
```
