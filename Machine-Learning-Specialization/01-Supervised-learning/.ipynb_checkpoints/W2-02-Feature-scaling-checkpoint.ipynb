```python
import numpy as np
import matplotlib.pyplot as plt

# Örnek veri
x = np.arange(0, 20, 1)  # 0'dan 19'a kadar sayılar
y = 3* x**2 + 2*x + 5  # Kuadratik bir fonksiyon

# Polinom özellikler (x, x^2, x^3)
X = np.c_[x, x**2, x**3]  # x, x^2, x^3 ekliyoruz
w = np.zeros(X.shape[1])  # Başlangıç parametreleri (0, 0, 0, 0)

# Gradient Descent Parametreleri
alpha = 0.0000001  # Öğrenme oranı (Küçük alınmalı, çünkü x^3 büyük değerler üretir)
epochs = 100000  # Güncelleme sayısı
m = len(y)

# Gradient Descent
for _ in range(epochs):
    predictions = np.dot(X,w)  # h_theta(x) hesapla
    errors = predictions - y  # Hata hesapla
    #print(f"errors: \n{errors.shape}")
    w -= (alpha / m) * (np.dot(X.transpose(),errors))  # Parametreleri güncelle
    #print(f"w: {w.shape}")

# Son parametreleri yazdır
print("Öğrenilen Parametreler:", w)

# Tahminleri çiz
plt.scatter(x, y, label="Gerçek Veriler")
plt.plot(x, np.dot(X,w), color='red', label="Model Tahmini")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
x = np.arange(0,30,1)
y = 4 * x ** 2 - 7 * x + 10
X = np.c_[x, x**2]
X_normalized = StandardScaler().fit_transform(X)

w = np.zeros(X.shape[1])
alpha = 0.0000001
iteration = 100000
m = x.shape[0]

for _ in range(iteration):
    prediction = np.dot(X, w)
    errors = prediction - y
    w -= (alpha / m) * (np.dot(X.transpose(), errors))


plt.scatter(x,y, label="gerçek veriler")
plt.plot(x, np.dot(X,w), c="r", label="tahminlerimiz")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
x = np.arange(0,30,1)
noise = np.random.normal(0, 30, size=x.shape) # tamamen şov
y = - 4 * x ** 2 + 7 * x - 10 + noise
X = np.c_[x, x**2]
X_normalized = StandardScaler().fit_transform(X)

w = np.zeros(X.shape[1])
alpha = 0.000001
iteration = 100000
m = x.shape[0]

for _ in range(iteration):
    prediction = np.dot(X, w)
    errors = prediction - y
    w -= (alpha / m) * (np.dot(X.transpose(), errors))


plt.scatter(x,y, label="gerçek veriler")
plt.plot(x, np.dot(X,w), c="r", label="tahminlerimiz")
```
